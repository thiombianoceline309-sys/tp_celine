### 1. Introduction

- Flask + Plotly + React => `Dash`

- La première version de `Dash` date de juin 2017.

- Il permet de développer des applications Web interactives de visualisation entièrement en Python

- Autre avantage de `Dash` est qu'on a accès à toutes les fonctionnalités graphiques de `plotly`

- De plus, vu qu'on utilise `Python`, on peut bien profiter des nombreuses librairies d'analyse et de traitement de données (Pandas, Scikit-Learn, ...)


## dcc = Dash Core Components

C’est un module de Dash qui contient les composants interactifs principaux :

Graphiques

Inputs

Dropdown

Slider

Upload

DatePicker

Tabs

etc.

## Différence entre html et dcc

Dans Dash il y a 2 modules principaux :

🔹 html

Pour la structure HTML :

html.Div == division

html.H1 == Titre de 1er niveau

html.P  ==  nouveua parapgraphe

html.Button == button interactif (clicable)

🔹 dcc

Pour les composants interactifs :

dcc.Graph

dcc.Dropdown == Pour faire le choix ou seclectionnner (cas selection des pays)

dcc.Input

dcc.Upload

dcc.Slider

#### Exemple 1 : Structure basique d'une application Dash

````python
# ======= Importation des librairies ========== #
import dash
from dash import html, dcc, Input, Output

# ============ Statistiques =============== #


# ===== Initialisation de l'application =========== #
app = dash.Dash()

# ==================== Layout ==================== #
app.layout = html.Div(children=[
        html.H1(children='Hello Dash!'),
    ]
)

# ==================== Callbacks ==================== #


# ==================== Run server ==================== #
if __name__ == '__main__':
    app.run(debug=True, port=8050, jupyter_mode="external")
````

#### Exemple 2 : Structure de base d'une application Dash avec un Input et un Output

````python
# ======= Importation des librairies ========== #
import dash
from dash import dcc, html, Input, Output, callback

# ===== Initialisation de l'application =========== #
app = dash.Dash()

# ==================== Layout (strure et interface de l'appliction)==================== #
app.layout = html.Div(children=[  #Le layout définit l’interface graphique./ Tout est contenu dans un html.Div(conteneur princapil de la page)
    html.H2(children='Hello Dash!'), ## Titre de la page (de 2ème niveau) //children = ce qui est affiché à l’intérieur
    
    dcc.Input(id='input-1', value='', type='text'),  # (C’est une zone où l’utilisateur peut écrire.) input de type texte avec une valeur par défaut vide
    html.Div(id='div-1', children='')                # zone d'affichage: output dans une div avec un id pour la callback
]) # id='div-1' → important pour la callback  /  children='' → contenu vide au départ

# ==================== Callbacks (interactivité)==================== #
@callback( #Quand la valeur de input-1 change → Mettre à jour le contenu (children) de div-1
    Output(component_id='div-1', component_property='children'), # (Modifier le contenu de div-1.)  Output : ce qui doit être mis à jour (div-1) / component_property='children' : ce qui doit être mis à jour dans div-1
    [Input(component_id='input-1', component_property='value')] # (Lire la valeur tapée dans le champ texte.)  Input : ce qui déclenche la mise à jour (input-1) / component_property='value' : la valeur de input-1 qui déclenche la mise à jour
)   
def update_div(input): # La fonction qui est appelée à chaque fois que la valeur de input-1 change. Elle prend en argument la nouvelle valeur de input-1 (input) et retourne une chaîne de caractères qui sera affichée dans div-1.
    return f'Texte introduit : {input}' # (Retourner une chaîne de caractères qui sera affichée dans div-1.)  La fonction retourne une chaîne de caractères qui inclut la valeur de input, et cette chaîne sera affichée dans div-1.

# ==================== Run server ==================== #
if __name__ == '__main__': # (Lancer le serveur Dash.)  Cette condition vérifie si le script est exécuté directement (plutôt que importé comme module), et si c’est le cas, elle lance le serveur Dash.
    app.run(debug=False, port=8050, jupyter_mode="external") # (Lancer le serveur Dash en mode debug sur le port 8050.)  app.run() : méthode pour lancer le serveur Dash / debug=False : désactive le mode debug / port=8050 : spécifie le port sur lequel le serveur écoute / jupyter_mode="external" : indique que l’application est exécutée dans un environnement Jupyter Notebook et que le serveur doit être lancé de manière à ce qu’il soit accessible depuis l’extérieur du notebook.
````

````python
from dash import Dash, html, dcc

app = Dash(__name__)

app.layout = html.Div([
    html.H1("Mon application Dash"),
    
    dcc.Dropdown( # Un menu déroulant pour sélectionner un pays
        options=[
            {"label": "France", "value": "FR"},
            {"label": "Burkina", "value": "BF"}
        ],
        value="FR"
    ),
    
    dcc.Graph(
        figure={
            "data": [
                {"x": [1,2,3], "y": [4,1,2], "type": "line"}
            ]
        }
    )
])

if __name__ == "__main__":
    app.run(debug=False, port=8057)
````
