## Dash Core

Dans ce chapitre nous abordons les composants `Input` de la librairie `dash_core_components`. Ces composants permettent de renseigner des champs de formulaire ou de recuperer des informations renseignées par l'utilisateur.

### Dropdown

Ce composant permet de proposer une liste de choix à l'utilisateur avec la possibilité de faire un choix mutiple.

````python
import dash
from dash import dcc, html

app = dash.Dash()
app.layout = html.Div([
    html.Label('Sélectionnez votre technologie préférée :'),

    dcc.Dropdown( # (Créer un composant Dropdown pour permettre à l'utilisateur de sélectionner une ou plusieurs technologies préférées. Le composant Dropdown est un élément interactif qui affiche une liste déroulante d'options parmi lesquelles l'utilisateur peut choisir.)
        options=[
            {'label': 'Python Exemple', 'value': 'py'},
            {'label': 'SQL', 'value': 'sql'},
            {'label': 'Java', 'value': 'jar'},
            {'label': 'R', 'value': 'r'},
        ],
        value='py',
        multi=True,
        searchable=True,
        placeholder='Choisissez la technologie...',
        disabled=False
    )
])

if __name__ == '__main__':
    app.run(debug=True, port=8056, jupyter_mode='external')
````

### Slider

Ce composant permet de choisir une valeur dans un intervalle.

````python
import dash
from dash import dcc, html
import plotly.graph_objects as go

app = dash.Dash()

app.layout = html.Div([

    html.Label('Slider'),
    dcc.Slider(min=0, max=3, step=1),

    dcc.Slider( # 
        min=0,
        max=3,
        step=1,
        marks={0: 'Label 0', 1: 'Label 1', 2:' Label 2', 3: 'Label 3'}
    ),

    html.Label("Range slider"),
    dcc.RangeSlider( # (Créer un composant RangeSlider pour permettre à l'utilisateur de sélectionner une plage de valeurs. Le composant RangeSlider affiche deux curseurs que l'utilisateur peut déplacer pour définir une plage de valeurs entre un minimum et un maximum spécifiés.)
        min=0,
        max=10,
        step=1,
    ),
])

if __name__ == '__main__':
    app.run(debug=True, port=8050, jupyter_mode='external')
````

### Input et textarea

Ce composant permet de rentrer du texte.

````python
import dash
from dash import dcc, html
import plotly.graph_objects as go

app = dash.Dash()

app.layout = html.Div([ 
    html.Div([
        html.H1("Input"),
        dcc.Input( type='text', placeholder='Entrez du texte...' ), # (Créer un composant Input pour permettre à l'utilisateur de saisir du texte. Le composant Input affiche une zone de saisie de texte standard, et le paramètre placeholder est utilisé pour afficher un texte d'instruction à l'intérieur de la zone de saisie lorsque celle-ci est vide.)
        dcc.Input( type='number', placeholder='Entrez un nombre...'),
        dcc.Input( type='password', placeholder='Entrez un mot de passe...'  ),
        dcc.Input( type='email', placeholder='Entrez un mail...' ),
    ]),

    html.Div([
        html.H1("Textarea"), 
        dcc.Textarea( # (Créer un composant Textarea pour permettre à l'utilisateur de saisir du texte sur plusieurs lignes. Le composant Textarea affiche une zone de texte plus grande que le composant Input standard, ce qui est utile pour les entrées de texte plus longues ou les commentaires.)
            placeholder='Entrez une valeur',
            style={'width': '100%'}, # (Définir le style du composant Textarea pour qu'il occupe toute la largeur disponible en utilisant le paramètre style avec une largeur de 100%. Cela permet au composant Textarea de s'adapter à la taille de son conteneur parent, offrant ainsi une expérience utilisateur plus fluide et une meilleure utilisation de l'espace.)
            value=''
        ), # style={'width': '100%'} pour que le composant Textarea occupe toute la largeur disponible, ce qui permet à l'utilisateur de saisir du texte sur plusieurs lignes sans être limité par la largeur du composant.
        dcc.Textarea(
            placeholder='Entrez une valeur',
            style={'width': '60%'} # (Définir le style du composant Textarea pour qu'il occupe 60% de la largeur disponible en utilisant le paramètre style avec une largeur de 60%. Cela permet au composant Textarea de s'adapter à la taille de son conteneur parent tout en laissant de l'espace pour d'autres éléments sur la même ligne ou à côté, offrant ainsi une mise en page plus équilibrée et une meilleure utilisation de l'espace.)
        ), # style={'width': '60%'} pour que le composant Textarea occupe 60% de la largeur disponible, ce qui permet à l'utilisateur de saisir du texte sur plusieurs lignes tout en laissant de l'espace pour d'autres éléments sur la même ligne ou à côté, offrant ainsi une mise en page plus équilibrée et une meilleure utilisation de l'espace. 
    ])
])

if __name__ == '__main__':
    app.run(debug=True, port=8050, jupyter_mode='external')
````

### Checklist et RadioItems

Ce composant permet de faire des choix multiples.

````python
import dash
from dash import dcc, html
import plotly.graph_objects as go

app = dash.Dash()

app.layout = html.Div([ 
    dcc.Checklist( # (Créer un composant Checklist pour permettre à l'utilisateur de sélectionner plusieurs options parmi une liste de choix. Le composant Checklist affiche une série de cases à cocher, et l'utilisateur peut cocher ou décocher les cases pour indiquer ses préférences ou ses sélections.)
        options=[
            {'label': 'Python', 'value': 'py'},
            {'label': 'SQL', 'value': 'sql'},
            {'label': 'Java', 'value': 'jar', 'disabled': True}
        ],
        value=['py', 'sql']
    ),
    dcc.RadioItems( # (Créer un composant RadioItems pour permettre à l'utilisateur de sélectionner une seule option parmi une liste de choix. Le composant RadioItems affiche une série de boutons radio, et l'utilisateur peut sélectionner un seul bouton pour indiquer sa préférence ou sa sélection.)
        options=[
            {'label': 'Python', 'value': 'py'},
            {'label': 'SQL', 'value': 'sql'},
            {'label': 'Java', 'value': 'jar', 'disabled': True} # (Ajouter une option "Java" au composant RadioItems avec une valeur de "jar" et la propriété disabled définie sur True pour désactiver cette option, ce qui signifie que l'utilisateur ne pourra pas la sélectionner. Cela peut être utilisé pour indiquer que l'option "Java" n'est pas disponible ou n'est pas applicable dans le contexte actuel.)
        ],
        value='py',
        labelStyle={'display': 'inline-block'} # (Définir le style des étiquettes des boutons radio pour les afficher en ligne en utilisant le paramètre labelStyle avec une propriété CSS display définie sur 'inline-block'. Cela permet aux étiquettes des boutons radio de s'afficher côte à côte plutôt que les unes en dessous des autres, offrant ainsi une présentation plus compacte et visuellement agréable.)
    ),

    html.Button('Sauvegarder'), # (Créer un composant Button pour permettre à l'utilisateur de déclencher une action, comme sauvegarder des données ou soumettre un formulaire. Le composant Button affiche un bouton cliquable avec le texte "Sauvegarder", et lorsque l'utilisateur clique sur ce bouton, une action peut être définie pour être exécutée, comme l'enregistrement des données saisies ou la soumission d'un formulaire.)
    html.Button('Sauvegarder', disabled=True),
    html.Button('Sauvegarder', type='submit'),
])

if __name__ == '__main__':
    app.run(debug=True, port=8056, jupyter_mode='external')
````

### Date picker

Ce composant permet de selectionner une date ou un intervalle

````python
import dash
from dash import dcc, html
from datetime import datetime as dt

app = dash.Dash()

app.layout = html.Div([
    
    html.H3("Simple date"),
    dcc.DatePickerSingle( # (Créer un composant DatePickerSingle pour permettre à l'utilisateur de sélectionner une date unique. Le composant DatePickerSingle affiche un calendrier interactif, et l'utilisateur peut cliquer sur une date pour la sélectionner. Le paramètre date est utilisé pour définir la date initiale affichée dans le calendrier.)
        date=dt(2019, 1, 1) 
    ),
    
    html.H3("Simple date avec format personnalisé"),
    dcc.DatePickerSingle(
        date=dt(2019, 1, 1),
        display_format='YYYY-MM-DD'
    ),

    html.H3("Plage de dates"),
    dcc.DatePickerRange(
        start_date=dt(2019, 9, 1),
        end_date=dt(2019, 10, 31)
    )
])

if __name__ == '__main__':
    app.run(debug=True, port=8050, jupyter_mode='external')
````

### Graph

Ce composant permet de tracer/afficher des graphiques

````python
import dash
from dash import dcc, html
import plotly.graph_objects as go

app = dash.Dash()

app.layout = html.Div([
    html.H3('Hello World - comparaison'),
    dcc.Graph(
        figure=go.Figure(
            data=[
                go.Bar( 
                    x=[2017, 2018, 2019], 
                    y=[219, 146, 167], 
                    name='Vente USA',
                    marker=go.bar.Marker( color='rgb(55, 83, 109)' ) # marker=go.bar.Marker( color='rgb(55, 83, 109)' ) pour définir la couleur des barres du graphique en utilisant une nuance de bleu foncé, ce qui permet de différencier visuellement les données de vente aux États-Unis des autres données présentées dans le même graphique.
                ),
                go.Bar(
                    x=[2017, 2018, 2019], 
                    y=[230, 180, 260], 
                    name='Vente Europa',
                    marker=go.bar.Marker( color='rgb(26, 118, 255)' ) # marker=go.bar.Marker( color='rgb(26, 118, 255)' ) pour définir la couleur des barres du graphique en utilisant une nuance de bleu plus claire, ce qui permet de différencier visuellement les données de vente en Europe des autres données présentées dans le même graphique.
                )
            ],
            layout=go.Layout( title='Vente', showlegend=True ) # layout=go.Layout( title='Vente', showlegend=True ) pour définir le titre du graphique à "Vente" et afficher la légende pour différencier les différentes séries de données présentées dans le graphique, ce qui permet aux utilisateurs de comprendre rapidement les informations représentées.
        )
    )
])

if __name__ == '__main__':
    app.run(debug=True, port=8050, jupyter_mode='external')
````

### Tabs

Ce composant permet de grouper les composants dans des onglets

````python
import dash
from dash import dcc, html

app = dash.Dash()

app.layout = html.Div([

    html.H3('Hello World - comparaison'),

    dcc.Tabs(children=[ # 
        dcc.Tab( label='Python', children=[
                dcc.Markdown(
                    """
                    Exemple de code Python :
                    ```python
                    print('Hello World')
                    ```
                    """
                )
        ]),
        dcc.Tab( label='Typescript', children=[
                dcc.Markdown(
                    """
                    Exemple de code Typescript :
                    ```ts
                    console.log('Hello World')
                    ```
                    """
                )
            ])
    ])
])

if __name__ == '__main__':
    app.run(debug=True, port=8050, jupyter_mode='external')
````

### Upload file

Ce composant permet de charger un fichier

````python
import dash
from dash import dcc, html

app = dash.Dash()

app.layout = html.Div([
    dcc.Upload(html.Button('Upload File')), # (Créer un composant Upload pour permettre à l'utilisateur de télécharger un fichier. Le composant Upload affiche un bouton cliquable avec le texte "Upload File", et lorsque l'utilisateur clique sur ce bouton, une fenêtre de sélection de fichier s'ouvre pour permettre à l'utilisateur de choisir un fichier à télécharger. Le composant Upload peut être configuré pour accepter différents types de fichiers en utilisant le paramètre accept, qui spécifie les types MIME des fichiers autorisés.)
    html.Hr(),

    dcc.Upload(html.A('Upload File'), accept='text/*'), # liste des mimetypes : https://mimetype.io/all-types
    html.Hr(),

    dcc.Upload([
        'Drag and drop or ',
        html.A('Select a File')
    ], style={
        'width': '100%',
        'height': '60px',
        'textAlign': 'center',
        'borderWidth': '1px',
        'borderStyle': 'dashed',
        'borderRadius': '5px',
        'lineHeight': '60px'
    })
])

if __name__ == '__main__':
    app.run(debug=True,  port=8050, jupyter_mode='external')
````
