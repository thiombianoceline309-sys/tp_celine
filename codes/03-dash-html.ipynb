### Application basique

````python
import dash
from dash import dcc, html
import plotly.graph_objects as go
 
app = dash.Dash()

app.layout = html.Div(children=[ #  Tout est contenu dans un html.Div(conteneur princapil de la page)
    html.H2(children='Hello Dash'), ## Titre de 2ème niveau
    html.Div(children='Juste une illustration de paragraphe'), ## Paragraphe

    dcc.Graph( # Un graphique avec des données fictives
        figure = go.Figure([
            go.Bar(
                x=[1, 2, 3],
                y=[2, 1, 3]
            )
        ])
    )
])

if __name__ == '__main__':
    app.run(debug=True, port=8053, jupyter_mode="external")
````

### Dash avec stylesheets

````python
import dash
from dash import dcc, html
import plotly.graph_objects as go

external_stylesheets = ["./assets/style.css"] # On peut aussi utiliser des stylesheets externes (ex: Bootstrap) en mettant l'URL à la place du chemin local
app = dash.Dash(external_stylesheets=external_stylesheets) # On peut aussi ajouter des styles CSS directement dans le code en utilisant l'argument style dans les composants Dash (ex: html.H2(style={'color': 'red'}))

app.layout = html.Div(children=[
    html.H2(children='Hello Dash', style={'color': '#b1f2c2'}),# Titre de 2ème niveau avec une couleur personnalisée
    html.Div(children='Juste une illustration de paragraphe', style={"color" : "brown", "font-size" : "bold"}), # Paragraphe avec une couleur et une taille de police personnalisées, bold pour mettre le texte en gras

    dcc.Graph(
        figure=go.Figure([
            go.Bar(
                x=[1, 2, 3],
                y=[2, 1, 3]
            )
        ])
    )
])

if __name__ == '__main__':
    app.run(debug=True, port=8053, jupyter_mode="external")
````

### Tableau de bord stylé

````python
import dash
from dash import dcc, html
import plotly.graph_objects as go # (Importer la bibliothèque plotly.graph_objects pour créer des graphiques plus personnalisés et complexes que ceux disponibles avec les composants de base de Dash.)

app = dash.Dash()

colors = {
    'background': '#b1f2c2',
    'text': '#4c524d'
}# (Creation d'un vecteur couleur)On peut aussi définir des styles CSS dans un dictionnaire Python et les appliquer à plusieurs composants en utilisant l’opérateur ** pour décompresser le dictionnaire dans les arguments de style.

app.layout = html.Div([
    html.H2('Hello Dash', style={'color': colors['text'], 'textAlign': 'center'}),
    html.Div('Dash: A web application framework for Python', 
             style={'color': colors['text'], 'textAlign': 'center'}),

    html.Div(children="Juste un test", style=colors),
    html.Div(children="Juste un deuxième test", style={**colors, "border" : "solid"}),#**colors pour appliquer les styles définis dans le dictionnaire colors, et "border" : "solid" pour ajouter une bordure solide autour du composant.

    dcc.Graph(
        figure=go.Figure([
            go.Bar( x=['2017', '2018', '2019'], y=[150, 180, 220],
                marker_color='#9ed6f0', marker_line_color='#4c524d',
                marker_line_width=5, 
                name='Ventes'
            ),# (Ajouter une trace de type Bar pour les ventes avec des couleurs personnalisées et un nom de légende.)
            go.Bar( x=['2017', '2018', '2019'], y=[160, 100, 280],
                    marker_color='#077eb5', marker_line_color='#4c524d',
                    marker_line_width=5, name='Achats'
            )], # (Ajouter une trace de type Bar pour les achats avec des couleurs personnalisées et un nom de légende.)
            layout=go.Layout( title='Visualisation des ventes et achats',
                              plot_bgcolor=colors['background'],
                              paper_bgcolor='#08825f'
            )# (Personnaliser le layout du graphique en utilisant les couleurs définies dans le dictionnaire colors pour le fond du graphique et du papier, et ajouter un titre au graphique.
        )
    )
    ],
    style={'backgroundColor': colors['background']} # (Appliquer une couleur de fond à toute la page en utilisant le style du conteneur principal html.Div
)

if __name__ == '__main__':
    app.run(debug=True, port=8053, jupyter_mode="external")
````

### Table classique

````python
import dash
from dash import html
import plotly.graph_objects as go

app = dash.Dash()

colors = { 
    'background': "#93e9aa",
    'text': '#4c524d'
}

app.layout = html.Div([
    html.H4("Cotations boursières d'Amazon"), # Titre de 4ème niveau pour le titre du tableau

    html.Table([
        # table header
        html.Tr([
            html.Th('Date'),
            html.Th('Open'),
            html.Th('High'),
            html.Th('Low'),
            html.Th('Close'),
            html.Th('Volume')
        ]),

        # table body
        html.Tr([
            html.Td('2019-09-01'),
            html.Td('100'),
            html.Td('102'),
            html.Td('98'),
            html.Td('100'),
            html.Td('150000')
        ]),
        html.Tr([
            html.Td('2019-09-02'),
            html.Td('101'),
            html.Td('103'),
            html.Td('98'),
            html.Td('100'),
            html.Td('140000')
        ])
    ]) 

], style={'backgroundColor': colors['background']})

if __name__ == '__main__' :
    app.run(debug=True, port=8053, jupyter_mode="external")
````

````python
import dash
from dash import html
import plotly.graph_objects as go
import pandas_datareader.data as web # For financial data

app = dash.Dash()

df = web.DataReader(name="AMZN", data_source='stooq') # Récupérer les données boursières d'Amazon depuis la source "stooq" en utilisant la fonction DataReader de la bibliothèque pandas_datareader.
df.reset_index(inplace=True) # Réinitialiser l'index du DataFrame pour que la colonne "Date" devienne une colonne normale au lieu de l'index, ce qui facilite l'affichage des données dans le tableau HTML.
df = df[:30]# Sélectionner les 30 premières lignes du DataFrame pour limiter le nombre de lignes affichées dans le tableau HTML.
min_val = min(len(df), 30) # Calculer la valeur minimale entre le nombre de lignes du DataFrame et 30 pour s'assurer que nous n'essayons pas d'afficher plus de lignes que ce qui est disponible dans le DataFrame, ce qui pourrait entraîner une erreur.

app.layout = html.Div([
    html.H4("Cotations boursières d'Amazon"),

    html.Table([
        html.Thead([ html.Th(col) for col in df.columns ])] + # (Créer l'en-tête de la table en utilisant les noms des colonnes du DataFrame df. La fonction html.Th est utilisée pour créer des cellules d'en-tête de table, et une compréhension de liste est utilisée pour itérer sur les noms des colonnes du DataFrame et créer une cellule d'en-tête pour chaque colonne.)
        [html.Tr([ html.Td(df.iloc[i][col]) for col in df.columns ]) for i in range(min_val)]
    )# (Créer une table HTML en utilisant les données du DataFrame df. La première partie de la table (html.Thead) crée l'en-tête de la table en utilisant les noms des colonnes du DataFrame, tandis que la deuxième partie (html.Tr) crée les lignes de la table en itérant sur les premières min_val lignes du DataFrame et en affichant les valeurs de chaque colonne pour chaque ligne.)
])

if __name__ == '__main__':
    app.run(debug=True, port=8053, jupyter_mode="external")
````

### Graphique

````python
import dash
from dash import dcc, html
import plotly.graph_objects as go
import pandas_datareader.data as web # For financial data 

df = web.DataReader(name="AMZN", data_source='stooq')
df.reset_index(inplace=True)
df = df[df.Date > '2019-01-01']

app = dash.Dash()

app.layout = html.Div([
    html.H4("Cotations boursières d'Amazon"),
    dcc.Graph(
        figure=go.Figure(
            data=[
                go.Scatter( x=df.Date, y=df.Close, mode='lines', fill='tozeroy',name='Amazon')
            ],
            layout=go.Layout( yaxis_type='log', height=500, title_text='Tableau des prix', showlegend=True)
        )# (Créer un graphique de type Scatter pour afficher les prix de clôture d'Amazon au fil du temps. Le graphique utilise une échelle logarithmique pour l'axe des y, une hauteur de 500 pixels, un titre "Tableau des prix", et affiche une légende pour la trace "Amazon".
    )
])

if __name__ == '__main__':
    app.run(debug=True, port=8053, jupyter_mode="external")
````

````python

import dash
from dash import dcc, html
import dash_bootstrap_components as dbc

app = dash.Dash(external_stylesheets=[dbc.themes.BOOTSTRAP]) # (Créer une application Dash en utilisant le thème Bootstrap pour les styles. L'argument external_stylesheets permet d'inclure des feuilles de style externes, et ici nous utilisons le thème Bootstrap fourni par la bibliothèque dash_bootstrap_components.)

app.layout = dbc.Container([ # (Utiliser le composant dbc.Container pour créer un conteneur principal pour la mise en page de la page. Le composant Container est un élément de base dans Bootstrap qui permet de centrer et de contenir le contenu de la page avec des marges appropriées.)
    dbc.Row([
        dbc.Col(children='Col 1 de taille 4', md=4, style={"border" : "solid"}, className="vh-60"), # classe "vh-60" pour donner à la première colonne une hauteur de 60% de la hauteur de la fenêtre, ce qui peut être utile pour démontrer l'utilisation de classes CSS personnalisées.
        dbc.Col(children='Col 2 de taille 4', md=4, style={"border" : "solid"}),# md=4 pour indiquer que la colonne doit occuper 4 des 12 colonnes disponibles dans la grille Bootstrap, et style={"border" : "solid"} pour ajouter une bordure solide autour de la colonne.
        dbc.Col(children='Col 3 de taille 4', md=4, style={"border" : "solid"}), 
    ], style={"height" : "150px"}), # (Créer une ligne avec trois colonnes de taille égale (md=4) en utilisant le composant dbc.Row pour la ligne et le composant dbc.Col pour les colonnes. Chaque colonne contient du texte et a une bordure solide pour les différencier visuellement. La ligne a une hauteur de 150 pixels définie dans le style.)
    dbc.Row(["Row 2"], style={"border" : "solid"}),
    dbc.Row([
        dbc.Col(children='Col 1 de taille 6', width=6, style={"border" : "solid"}),
        dbc.Col(children='Col 2 de taille 6', width=6, style={"border" : "solid"}),
    ]),
    dbc.Row([
        dbc.Col(children='Col 1 de taille 8', width=8, style={"border" : "solid"}),
        dbc.Col(children='Col 2 de taille 2', width=2, style={"border" : "solid"}),
        dbc.Col(children='Col 3 de taille 2', width=2, style={"border" : "solid"}),
    ]),
], fluid=True)

if __name__ == '__main__': 
    app.run(debug=True, port=8052, jupyter_mode="external")
````

### Dash Bootstrap exemple

````python

import dash
from dash import dcc, html
import plotly.graph_objects as go
import dash_bootstrap_components as dbc

app = dash.Dash(external_stylesheets=[dbc.themes.BOOTSTRAP])

graph = dcc.Graph(figure={
    'data': [
        {'x': [1, 2, 3], 'y': [4, 1, 2], 'type': 'bar', 'name': 'SF'},
        {'x': [1, 2, 3], 'y': [2, 4, 5], 'type': 'bar', 'name': 'Montréal'},
    ],
    'layout': { 'title': 'Dash Data Visualization'},
}, responsive=True, className="h-100")

app.layout = dbc.Container([  # (Utiliser le composant dbc.Container pour créer un conteneur principal pour la mise en page de la page. Le composant Container est un élément de base dans Bootstrap qui permet de centrer et de contenir le contenu de la page avec des marges appropriées.)
    dbc.Row([
        dbc.Col(html.Div("stuff", className="bg-secondary h-100"), width=2),
        dbc.Col([
            dbc.Row([ dbc.Col([graph]) ]),
            dbc.Row([ dbc.Col([graph]) ]),
        ], width=5),
        dbc.Col([
            dbc.Row([ dbc.Col([graph]) ], className="h-100"),
        ], width=5),
    ])
], fluid=True)

if __name__ == '__main__': 
    app.run(debug=True, port=8052, jupyter_mode="external")
````
