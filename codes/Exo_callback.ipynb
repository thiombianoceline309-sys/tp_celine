```python
import pandas as pd
import plotly.express as px
import dash
from dash import dcc, html, Input, Output, callback

df = pd.read_csv("./datasets/data.csv")

list_categories = []
for cat in df['Product_Category'].unique():
    list_categories.append({'label': cat, 'value': cat})
    
list_categories = [{'label': cat, 'value': cat} for cat in df['Product_Category'].unique()]

app = dash.Dash()
app.layout = html.Div([
    dcc.Dropdown(
        id='drop-1',
        options=list_categories,
        value=list_categories[0]['value'],
        searchable=True
    ),
    dcc.Graph(id='graph-1')
])

@callback(
    Output('graph-1', 'figure'),
    [Input('drop-1', 'value')]
)
def update_graph(value):
    if value is None:
        df_filter = df
    else:
        df_filter = df[df['Product_Category'] == value]
    
    df_plot = df_filter.groupby('Gender').size().reset_index(name='count')
    fig = px.bar(df_plot, x="Gender", y="count")
    return fig

if __name__ == '__main__' :
    app.run(debug=True, port=8050, jupyter_mode='external')
```
